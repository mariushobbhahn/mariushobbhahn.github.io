```python
import numpy as np
import matplotlib.pyplot as plt

#### numbers

x = np.linspace(0,10, 500+1)


y_st_0 = 0.3*np.sin(x)
y_lt_0 = np.sin(5*x) 
y_lt_plot_0 = y_st_0 + y_lt_0

y_st_1 = np.sin(x)
y_lt_1 = 0.4* np.sin(5*x) 
y_lt_plot_1 = y_st_1 + y_lt_1

#### plot

fig, axs = plt.subplots(nrows=1, ncols=2, sharex=False, figsize=(23, 6))

axs[0].plot(x, y_st_0, label="long-term", linewidth=5)
axs[0].plot(x, y_lt_plot_0, label="short-term", linewidth=5)
axs[0].set_yticklabels([])
axs[0].set_yticks([])
axs[0].set_ylabel('Happiness', fontdict={'fontsize':25})
axs[0].set_title('Old View', fontdict={'fontsize': 30})
axs[0].legend(prop={"size":25})

axs[1].plot(x, y_st_1, label="long-term", linewidth=5)
axs[1].plot(x, y_lt_plot_1, label="short-term", linewidth=5)
axs[1].set_yticklabels([])
axs[1].set_yticks([])
axs[1].set_ylabel('Happiness', fontdict={'fontsize':25})
axs[1].set_title('New View', fontdict={'fontsize': 30})
axs[1].legend( prop={"size":25})
plt.plot();

plt.tight_layout()
plt.savefig("emotions_influence.png", dpi=300)
```
