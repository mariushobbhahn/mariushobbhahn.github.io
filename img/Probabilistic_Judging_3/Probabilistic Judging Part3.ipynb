# Clashes: same metric

```python
from scipy.stats import beta, gamma

x_labels = ['', 'nothing happens', 'who cares', 'maybe something', 'actually something', 'big impact', 'nuclear war' ]

#relevance 

x_gamma_sm = np.linspace(-70, 70, 250+1)

s_sm= 1
#scenario 1: gov wins
g_g1_sm, g_o1_sm = 20, 40
gamma_g1_sm = gamma.pdf(x_gamma_sm, g_g1_sm, scale=s_sm)
gamma_o1_sm = gamma.pdf(x_gamma_sm, g_o1_sm, scale=s_sm)

#scenario 2: opp wins
g_g2_sm, g_o2_sm = 30, 10
gamma_g2_sm = gamma.pdf(x_gamma_sm, g_g2_sm, scale=s_sm)
gamma_o2_sm = gamma.pdf(x_gamma_sm, g_o2_sm, scale=s_sm)

#scenario 3: tie
g_g3_sm, g_o3_sm = 25, 25
gamma_g3_sm = gamma.pdf(x_gamma_sm, g_g3_sm, scale=s_sm)
gamma_o3_sm = gamma.pdf(x_gamma_sm, g_o3_sm, scale=s_sm)

#reverse the opposition
gamma_o1_sm = np.flip(gamma_o1_sm)
gamma_o2_sm = np.flip(gamma_o2_sm)
gamma_o3_sm = np.flip(gamma_o3_sm)


fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(30, 6))





axs[0].plot(x_gamma_sm, gamma_g1_sm, label='Gov argument')
axs[0].plot(x_gamma_sm, gamma_o1_sm, label='Opp argument')
axs[0].vlines(g_g1_sm * s_sm, ymin=0, ymax=1*gamma_g1_sm.max(), linestyle='dashed', alpha=1, color="#1f77b4")
axs[0].vlines(-g_o1_sm * s_sm, ymin=0, ymax=1*gamma_o1_sm.max(), linestyle='dashed', alpha=1, color="#ff7f0e")
axs[0].set_yticklabels([])
axs[0].set_yticks([])
axs[0].set_xlabel('metric', fontsize=25)
axs[0].legend(prop={'size': 22})
axs[0].set_title('Scenario 1', fontdict={'fontsize': 25})

axs[1].plot(x_gamma_sm, gamma_g2_sm, label='Gov argument')
axs[1].plot(x_gamma_sm, gamma_o2_sm, label='Opp argument')
axs[1].vlines(g_g2_sm * s_sm, ymin=0, ymax=1*gamma_g2_sm.max(), linestyle='dashed', alpha=1, color="#1f77b4")
axs[1].vlines(-g_o2_sm * s_sm, ymin=0, ymax=1*gamma_o2_sm.max(), linestyle='dashed', alpha=1, color="#ff7f0e")
axs[1].set_yticklabels([])
axs[1].set_yticks([])
axs[1].set_xlabel('metric', fontsize=25)
axs[1].legend(prop={'size': 22})
axs[1].set_title('Scenario 2', fontdict={'fontsize': 25})

axs[2].plot(x_gamma_sm, gamma_g3_sm, label='Gov argument')
axs[2].plot(x_gamma_sm, gamma_o3_sm, label='Opp argument')
axs[2].vlines(g_g3_sm * s_sm, ymin=0, ymax=1*gamma_g3_sm.max(), linestyle='dashed', alpha=1, color="#1f77b4")
axs[2].vlines(-g_o3_sm * s_sm, ymin=0, ymax=1*gamma_o3_sm.max(), linestyle='dashed', alpha=1, color="#ff7f0e")
axs[2].set_yticklabels([])
axs[2].set_yticks([])
axs[2].set_xlabel('metric', fontsize=25)
axs[2].legend(prop={'size': 22})
axs[2].set_title('Scenario 3', fontdict={'fontsize': 25})

for ax in fig.axes:
    plt.sca(ax)
    plt.xticks(fontsize = 20)

plt.tight_layout()
plt.savefig('same_metric.png', dpi=300)

plt.show();
```

```python
#2D meta prior

#create some distributions
from scipy.stats import beta, gamma

x = np.linspace(0,1, 100 +1)

#meta prior 1
a_mp1, b_mp1 = 4,6
beta_mp1 = beta.pdf(x, a_mp1, b_mp1)
beta_mp1_exp = a_mp1/(a_mp1+b_mp1)

#meta prior 2
a_mp2, b_mp2 = 6, 0.7
beta_mp2 = beta.pdf(x, a_mp2, b_mp2)
beta_mp2_exp = a_mp2/(a_mp2+b_mp2)

#meta prior 3
a_mp3, b_mp3 = 1.5, 1.5
beta_mp3 = beta.pdf(x, a_mp3, b_mp3)
beta_mp3_exp = a_mp3/(a_mp3+b_mp3)


fs_tick_labels = 20
fig, axs = plt.subplots(nrows=1, ncols=3, sharex=False, figsize=(23, 6))

axs[0].set_title('Scenario 1', fontdict={'fontsize': 25})
axs[0].plot(x, beta_mp1, label='meta prior 1')
axs[0].vlines(beta_mp1_exp, ymin=0, ymax=1*beta_mp1.max(), linestyle='dashed', alpha=1, color="#1f77b4")
axs[0].set_yticklabels([])
axs[0].set_xticklabels(['', 'fairness', '','', '', '', 'utility'], fontdict={'fontsize':fs_tick_labels})
axs[0].set_yticks([])
axs[0].legend(prop={'size': 25})

axs[1].set_title('Scenario 2', fontdict={'fontsize': 25})
axs[1].plot(x, beta_mp2, label='meta prior 2')
axs[1].vlines(beta_mp2_exp, ymin=0, ymax=1*10, linestyle='dashed', alpha=1, color="#1f77b4")
axs[1].set_yticklabels([])
axs[1].set_yticks([])
axs[1].set_xticklabels(['', 'rights', '','', '', '', '#deaths'], fontdict={'fontsize':fs_tick_labels})
axs[1].legend(prop={'size': 25})

axs[2].set_title('Scenario 3', fontdict={'fontsize': 25})
axs[2].plot(x, beta_mp3, label='meta prior 3')
axs[2].vlines(beta_mp3_exp, ymin=0, ymax=1*1.3, linestyle='dashed', alpha=1, color="#1f77b4")
axs[2].set_yticklabels([])
axs[2].set_yticks([])
axs[2].set_xticklabels(['', 'rights', '', '', '', '', 'utility'], fontdict={'fontsize':fs_tick_labels})
axs[2].legend(prop={'size': 25})

for ax in fig.axes:
    plt.sca(ax)
    plt.xticks(rotation=18)

plt.tight_layout()
plt.savefig('2D_meta_priors.png', dpi=300)

plt.show();
```

```python
#3D meta prior

# just copy pasted from this link: http://blog.bogatron.net/blog/2014/02/02/visualizing-dirichlet-distributions/
%matplotlib inline
import numpy as np
import matplotlib.tri as tri
import math
from functools import reduce
from math import gamma
from operator import mul
from scipy.special import loggamma, gammaln, xlogy

corners = np.array([[0, 0], [1, 0], [0.5, 0.75**0.5]])
triangle = tri.Triangulation(corners[:, 0], corners[:, 1])

refiner = tri.UniformTriRefiner(triangle)
trimesh = refiner.refine_triangulation(subdiv=4)


midpoints = [(corners[(i + 1) % 3] + corners[(i + 2) % 3]) / 2.0 \
             for i in range(3)]
def xy2bc(xy, tol=1.e-3):
    '''Converts 2D Cartesian coordinates to barycentric.'''
    s = [(corners[i] - midpoints[i]).dot(xy - midpoints[i]) / 0.75 \
         for i in range(3)]
    return np.clip(s, tol, 1.0 - tol)

def logit_transform(x):
    
    out = np.log(x/(1-x))
    return(out)

def beta_function(alpha):
    return(np.exp(np.sum([loggamma(a_i) for a_i in alpha]) - loggamma(np.sum(alpha))))    

class Dirichlet(object):
    def __init__(self, alpha):
        self._alpha = np.array(alpha)
        self._coef = np.sum(gammaln(alpha)) - gammaln(np.sum(alpha))    
        
    def pdf(self, x):
        '''Returns pdf value for `x`.'''
        #x = softmax_transform(x)
        return(np.exp(- self._coef+ np.sum((xlogy(self._alpha-1, x.T)).T, 0)))

def draw_pdf_contours(dist, nlevels=200, subdiv=4, filename='test.png',**kwargs):

    refiner = tri.UniformTriRefiner(triangle)
    trimesh = refiner.refine_triangulation(subdiv=subdiv)
    pvals = [dist.pdf(xy2bc(xy)) for xy in zip(trimesh.x, trimesh.y)]
    xys = [xy2bc(xy) for xy in zip(trimesh.x, trimesh.y)]
    #print(len(pvals))
    #print(xys)
    print(len(xys))

    cs = plt.tricontourf(trimesh, pvals, nlevels, cmap=new_coolwarm) #plasma, inferno is OK
    for c in cs.collections:
        c.set_rasterized(True)
    plt.axis('equal')
    plt.xlim(0, 1)
    plt.ylim(0, 0.75**0.5)
    plt.axis('off')
    plt.tight_layout()
    plt.margins(0,0)
    plt.gca().xaxis.set_major_locator(plt.NullLocator())
    plt.gca().yaxis.set_major_locator(plt.NullLocator())
    plt.savefig('{}'.format(filename), bbox_inches = 'tight', pad_inches = 0)
    
draw_pdf_contours(Dirichlet([3,3,3]), filename='new_coolwarm.png')
```
